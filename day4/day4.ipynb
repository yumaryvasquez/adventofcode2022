#### Part 1

```python
#file = open("input_assignment.txt", 'r').read().splitlines()
file = open("input_assignment.txt", 'r')
```

```python
file=file.read().splitlines()
file=[sub.replace('-',',') for sub in file]
nested_list =[sub.split(",") for sub in file]

count = 0
for sublist in nested_list:
    if((set(range(int(sublist[0]),int(sublist[1])+1)).issubset(range(int(sublist[2]),int(sublist[3])+1)) == True) | (set(range(int(sublist[2]),int(sublist[3])+1)).issubset(range(int(sublist[0]),int(sublist[1])+1)) == True)):
        count += 1
count
```

#### Part 2

```python
count = 0
for sublist in nested_list:
    if((set(range(int(sublist[0]),int(sublist[1])+1)).intersection(range(int(sublist[2]),int(sublist[3])+1)) == True) | (set(range(int(sublist[2]),int(sublist[3])+1)).intersection(range(int(sublist[0]),int(sublist[1])+1)) == True)):
        count += 1
```

```python
count = 0 
for sublist in nested_list:
    s = set(range(int(sublist[0]),int(sublist[1])+1))
    if (any(e in s for e in range(int(sublist[2]),int(sublist[3])+1)) == True):
        count += 1
```
