#### Part 1

```python
#start time 4:15

#open file
rucksacks_items = open("input_item.txt", 'r').read().splitlines()

sum_num = 0
for comp in rucksacks_items:
    first = comp[:len(comp)//2] #grabs first half of sentence 
    second = comp[len(comp)//2:] #grabs first half of sentence 
    sameLetter = set(first).intersection(second) # finds the intersection
    listLetter = list(sameLetter) #changes into a list 
    for letter in listLetter: 
        if letter == letter.lower():
            low_letter = ord(letter)- 96  #takes ASCII number and makes a-z into 1-27
            sum_num += low_letter  #adds number to sum
        if letter == letter.upper(): #takes ASCII number and makes A-Z into 28 - 56
            upper_letter = ord(letter)- 38
            sum_num += upper_letter #adds number to sum
sum_num
    #print(ord(listLetter))
    #print(ord(sameLetter))
# 4:45 
#back at 5:05 pm
#done at 5:08 pm


```
